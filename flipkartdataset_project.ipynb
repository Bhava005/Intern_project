```python
#FILL MISIING VALUES BY PREVIOUS VALUE
import pandas as pd
data=pd.read_csv("flipkart.csv")
df=pd.DataFrame(data)
print(df.fillna(method='pad'))
```

```python
#FILL MISIING VALUES BY NEXT VALUE
import pandas as pd

data = pd.read_csv("flipkart.csv")
df = pd.DataFrame(data)
df_filled = df.bfill()
print(df_filled.to_string())
```

```python
#INTERPOLATE MISSING VALUES USING LINEAR METHOD
import pandas as pd
data = pd.read_csv("flipkart.csv")
df = pd.DataFrame(data)
df_interpolate=df.interpolate(method='linear',limit_direction='forward')
print(df_interpolate)
```

```python
#FILL A NULL VALUES USING replace() METHOD
import pandas as pd
import numpy as np
data=pd.read_csv("flipkart.csv")
data.replace(to_replace=np.nan,value=-999)
```

```python
#DROPPING ROWS WITH ATLEAST 1 NULL VALUE IN CSV FILE(along rows)
import pandas as pd
data=pd.read_csv("flipkart.csv")
new_data=data.dropna(axis=0,how='any')
new_data#resultant rows does not contains even a single null value
```

```python
#DROPPING ROWS WITH ATLEAST 1 NULL VALUE IN CSV FILE(along columns)
import pandas as pd
data=pd.read_csv("flipkart.csv")
new_data=data.dropna(axis=1,how='any')
new_data#resultant columns does not contains any null values
```

```python
#TO KNOW HOW MANY ROWS HAVE A SINGLE NULL VALUE
import pandas as pd
data=pd.read_csv("flipkart.csv")
new_data=data.dropna(axis=0,how='any')
print("Old dataframe length:",len(data))
print("New dataframe length:",len(new_data))
print("Number of rows with atleast 1 Null (NA) value:",len(data)-len(new_data))
```
